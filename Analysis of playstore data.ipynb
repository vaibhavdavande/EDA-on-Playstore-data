```python
#Load the required libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the dataset
ps = pd.read_csv('googleplaystore.csv')
ps.head()
```

```python
#Get the details of dataset its columns and datatypes etc
ps.info()
```

```python
#Lets find some unique values from Install column
ps.Installs.unique()
```

```python
#Lets drop unwanted contents
ps=ps.drop(ps[ps.Installs=='Free'].index, axis = 0)#deleted row from dataset which has a 'Free' entry
```

```python
#Data preprocessing 
ps.Installs = ps.Installs.str.replace('+','').str.replace(',','').astype('float')
```

```python
#Check datasets first 5 entries
ps.head()
```

```python
#Lets findout category wise distribution of softwares installed by the users 
ps[ps.Installs==ps.Installs.max()]['Category'].value_counts().plot.bar()
plt.title('Category Vs No.of users', fontdict = {'fontsize':15})
plt.show()
```
